# Simple Linear Regression

## steps:
- load the dataset
- specify indenpendent variable x
- fit the data from the dataset i.e create intercept
- specify dependent var
- create the model
- train the model
-

```python
# datasets Loading
from ISLP import load_data
dataset = load_data('Boston')
dataset.head(5)
```

```python
#specify X1
from ISLP.models import ModelSpec as MS
x1_design = MS(['lstat'])
X = x1_design.fit_transform(dataset) # create intercept values
X
```

```python
#create the dependendent variable / y
y = dataset['medv']
y.head(5)
```

```python
# train the linear regression model
import statsmodels.api as sm
model = sm.OLS(y,X) # initalize the model
res = model.fit() # train the model
```

```python

summarize(res)
```

```python
res.summary()
```

Multiple Linear Regession

```python
X2 = MS(['lstat' , 'age']).fit_transform(dataset)
model2 = sm.OLS(y,X2)
res2 = model2.fit()
summarize(res2)
```

```python
#test the model
import numpy as np
import pandas as pd

new_data = pd.DataFrame({'lstat' : [5 , 10 ,15],
                          'age' : [4 , 8 , 16]})
new_data_design = MS(new_data)
new_data = new_data_design.fit_transform(new_data)
predictions = res2.get_prediction(new_data)
predictions.predicted_mean
predictions.conf_int()
```

Interaction Terms

```python
X3 = MS(['lstat' , 'age' , ('lstat' , 'age')]).fit_transform(dataset)
model3 = sm.OLS(y , X3)
summarize(model3.fit())
```
