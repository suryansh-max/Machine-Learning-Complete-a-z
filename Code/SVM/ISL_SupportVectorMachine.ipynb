```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import subplots
import sklearn.model_selection as skm
from ISLP import confusion_table, load_data
```

```python
from sklearn.svm import SVC # this is support vector classifer to use svms
from ISLP.svm import plot # helps in plottings the smv models
```

# Generates Random Data

```python
random_linear = np.random.default_rng(1) # to generate linear data model
X = random_linear.standard_normal((100, 2)) # generate 100 random samples with 2 features x1,x2
y = np.array([0]*50 + [1]*50) # generate 100 deoendent var 50 each 1/0
X[y == 1] += 1
```

```python
#plot to make sure it is there
fig, ax = subplots(figsize=(5,5))
ax.scatter(X[:,0],X[:,1], c=y )
```

# Support Vector Classifier

```python
svc_model = SVC(kernel='linear' , C=15) # create model for svc, if we increase c region shrinks and vice versa
svc_model.fit(X,y) # train the model
y_pred = svc_model.predict(X) # get some predictions
print(confusion_table(y_pred , y)) # ct for this

fig, ax = subplots(figsize=(5,5))
plot(X,y,svc_model,ax=ax)
```

# Support vector Machines

```python
# Radial model
rad_model = SVC(kernel='rbf' , C=1) # create model(radial this time) for svc, if we increase c region shrinks and vice versa
rad_model.fit(X,y) # train the model
y_pred = rad_model.predict(X) # get some predictions
print(confusion_table(y_pred , y)) # ct for this

fig, ax = subplots(figsize=(5,5))
plot(X,y,rad_model,ax=ax)
```

Non linear Data

```python
random_linear = np.random.default_rng(1) # to generate linear data model
X = random_linear.standard_normal((300, 2)) # generate 100 random samples with 2 features x1,x2
X[:150] += 2
X[150:200] -= 2

y = np.array([0]*200 + [1]*100) # generate 100 deoendent var 50 each 1/0
# X[y == 1] += 1

#plot to make sure it is there
fig, ax = subplots(figsize=(5,5))
ax.scatter(X[:,0],X[:,1], c=y )
```

```python
X_train,X_test,y_train,y_test = skm.train_test_split(X, y, test_size=0.2)

# Radial model
rad_model = SVC(kernel='rbf' , C=1) # create model(radial this time) for svc, if we increase c region shrinks and vice  , if c values is low the model consider only 1 feature
rad_model.fit(X_train,y_train) # train the model
y_pred = rad_model.predict(X_test) # get some predictions
print(confusion_table(y_pred , y_test)) # ct for this

fig, ax = subplots(figsize=(5,5))
plot(X_train,y_train,rad_model,ax=ax)

# we can do principle component analysis if the feature size is too much to get pc1 and pc2
```
